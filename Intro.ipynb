```python
import pandas
import numpy
import random
import pygame
from collections import deque
# pygame setup
ancho=1000
alto=1000
pygame.init()
screen = pygame.display.set_mode((ancho, alto))
clock = pygame.time.Clock()
running = True
celdas=23
tablero=[]
comida = False
direccion=["der","izq","aba", "arr"]
adonde=""
moviendome = False
sitio=[0,0]
nsitio=[0,0]
ppunto=[0,0]
# Crear una cola vaca
fila = deque()



pmatriz=[0,0]

vel = 1


def matriz():
    matriz=[]
    for i in range(celdas+2):
        temp=[]
        if i==0 or i==celdas+1:
            for j in range(celdas+2):
                temp.append(-100) 
        else:
            for j in range(celdas+2):
                if j==0 or j==celdas+1:
                    temp.append(-100)
                else:
                    temp.append(-1)
        matriz.append(temp)
    
    numero_aleatoriox = random.randint(1, celdas)
    numero_aleatorioy = random.randint(1, celdas)
    matriz[numero_aleatorioy][numero_aleatoriox]=-50
    fila.append([numero_aleatoriox,numero_aleatorioy])
    return matriz, numero_aleatoriox, numero_aleatorioy,fila


def escenario():
    boleano = comida
    while (not comida):
        numero_aleatoriox = random.randint(1, celdas)
        numero_aleatorioy = random.randint(1, celdas)
        if tablero[numero_aleatorioy][numero_aleatoriox]!=-50:
            tablero[numero_aleatorioy][numero_aleatoriox]=100
            ppunto[1]=numero_aleatorioy
            ppunto[0]=numero_aleatoriox
            boleano = True
            break


    color = "black"
    thickness = 5
    for i in range(celdas):
        temporal=i*alto/celdas
        start_pos = (0, temporal)
        end_pos = (ancho, temporal)
        pygame.draw.line(screen, color, start_pos, end_pos, thickness)

        temporal=i*ancho/celdas
        start_pos = (temporal, 0)
        end_pos = (temporal, alto)
        pygame.draw.line(screen, color, start_pos, end_pos, thickness)

    matriz=[]
    for i in range(celdas+2):
        temp=[]
        if i==0 or i==celdas+1:
            for j in range(celdas+2):
                temp.append(-100) 
        else:
            for j in range(celdas+2):
                if j==0 or j==celdas+1:
                    temp.append(-100)
                else:
                    temp.append(-1)
        matriz.append(temp)

    tablero1=matriz

    for i in fila:
        tablero1[i[1]][i[0]]=-50
        if i == fila[-1]:
            tablero1[i[1]][i[0]]=0

    tablero1[ ppunto[1]][ppunto[0]]=100


    for i in range(len(tablero1)):
        for j in range(len(tablero1)):
            if tablero1[j][i]==100:
                pygame.draw.circle(screen, "red",( ancho/celdas*(i-1) + ancho/celdas/2 , alto/celdas*(j-1) + alto/celdas/2), 20)
            if tablero1[j][i]==-50:
                pygame.draw.circle(screen, "green",( ancho/celdas*(i-1) + ancho/celdas/2 , alto/celdas*(j-1) + alto/celdas/2), 20)
            if tablero1[j][i]==0:
                pygame.draw.circle(screen, "black",( ancho/celdas*(i-1) + ancho/celdas/2 , alto/celdas*(j-1) + alto/celdas/2), 20)

    return boleano , tablero, ppunto, tablero1


def move():
    x=sitio[1]
    y=sitio[0]
    nx = nsitio[1] 
    ny = nsitio[0]
    dir = adonde
    boleano1= moviendome
    com=True
    if not moviendome or not boleano1:
        numero_aleatorio = random.randint(0, 3)
 
        if direccion[numero_aleatorio] == "der" and x<len(tablero)-2 and tablero[y][x+1]!=-50:
            dir="der"
            boleano1= True
            nx= x+1

        if direccion[numero_aleatorio] == "izq" and x>1 and tablero[y][x-1]!=-50:
            dir="izq"
            boleano1= True
            nx= x-1

        if direccion[numero_aleatorio] == "aba" and y<len(tablero)-2 and tablero[y+1][x]!=-50:
            dir="aba"
            boleano1= True
            ny= y+1

        if direccion[numero_aleatorio] =="arr" and y > 1  and tablero[y-1][x]!=-50:
            dir="arr"
            boleano1= True
            ny= y-1 

    else:
        if dir == "der":
            x=x+vel

        if dir == "izq":
            x=x-vel
        
        if dir == "aba":
            y=y+vel
        
        if dir == "arr":
            y=y-vel
            


    if dir == "der":
        if (x>nx):
            boleano1=False
            tablero[y][nx-1]=-1
            x=nx
            if tablero[y][x]==100:
                com=False
                fila.append([x,y])
            else:
                fila.append([x,y])
                fila.popleft()

            

    if dir == "izq":
        if (x<nx):
            boleano1=False
            tablero[y][nx+1]=-1
            x=nx
            if tablero[y][x]==100:
                com=False
                fila.append([x,y])
            else:
                fila.append([x,y])
                fila.popleft()

    
    if dir == "aba":
        if (y>ny):
            boleano1=False
            tablero[ny-1][x]=-1
            y=ny
            if tablero[y][x]==100:
                com=False
                fila.append([x,y])
            else:
                fila.append([x,y])
                fila.popleft()
 

    if dir == "arr":
        if (y<ny):
            boleano1=False
            tablero[ny+1][x]=-1
            y=ny
            if tablero[y][x]==100:
                com=False
                fila.append([x,y])
            else:
                fila.append([x,y])
                fila.popleft()

    

    


    



    

        

    return boleano1,dir, y ,x , ny, nx, tablero, com, fila

tablero, sitio[0], sitio[1], fila=matriz()
pmatriz[0]=sitio[0]
pmatriz[1]=sitio[1]


while running:
    # poll for events
    # pygame.QUIT event means the user clicked X to close your window
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # fill the screen with a color to wipe away anything from last frame
    screen.fill((255, 253, 208))

    # RENDER YOUR GAME HERE




    comida, tablero, ppunto, tablero=escenario()
    moviendome, adonde, sitio[0], sitio[1], nsitio[0], nsitio[1], tablero, comida, fila= move()
    
    #pygame.draw.circle(screen, "red", (ancho/2, alto/2), 20)


    

    # Definir el grosor de la lnea (en este caso, 5 pxeles)
    

    # Dibujar la lnea en la ventana

    


    # flip() the display to put your work on screen
    pygame.display.flip()

    clock.tick(100)  # limits FPS to 60

pygame.quit()
```
